```python
import pandas as pd

url = "https://drive.google.com/uc?id=1XVKLHBpZRY4sdiKLC_Ip-VXEZpGKsFvS"
df = pd.read_csv(url)
df
```

```python
for col in df.columns:
    df[col].fillna(df[col].mean(),inplace=True)
df
```

```python

if df.isna().any().any():
    print("data frame has nan value")
else:
    print("no nan value remained")
```

```python
df.head(5)
```

```python
corr = df[["Glucose", "BloodPressure", "BMI", "Age", "Insulin", "Outcome"]].corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr, annot= True, cmap='coolwarm')
plt.title("Correlation heatmap")
plt.show()
```

```python
droped_corr = corr.drop("Outcome")
droped_corr
```

```python
sns.heatmap(droped_corr, annot=True, cmap='coolwarm')
plt.title("Correlation heatmap")
plt.show()
```
